```python
import matplotlib.pyplot as plt 
import numpy as np
from PIL import Image,ImageEnhance

rgb_to_hex = lambda rgb_ary:"#{:0>2X}{:0>2X}{:0>2X}".format(rgb_ary[0],rgb_ary[1],rgb_ary[2])
```

```python
file_path = r"./image/25c31d8d-21a4-9ea1-1d73-6a2eca4dda7e.jpg"
target_size = (80,60)
```

```python
img = Image.open(file_path).convert('RGB')

converter = ImageEnhance.Color(img)
new_img = converter.enhance(1.6)

new_img = new_img.resize(target_size,Image.NEAREST) # Image.NEAREST Image.BOX Image.BILINEAR Image.HAMMING Image.BICUBIC Image.LANCZOS
new_img_ary = np.asarray(new_img)
img.size
```

```python
# TODO : automatic point size
size = 25

fig,axes = plt.subplots(1,2,figsize=(20,15))

ax=axes[0]
ax.set_aspect('equal', adjustable='box')
ax.imshow(img, interpolation='nearest')
ax.axis('off')

ax=axes[1]
ax.set_aspect('equal', adjustable='box')
ax.scatter([i for i in range(new_img_ary.shape[1])] * new_img_ary.shape[0],
           [[i]*new_img_ary.shape[1] for i in range(new_img_ary.shape[0]-1,0-1,-1)],
           c=list(map(rgb_to_hex,new_img_ary.reshape(-1,3))),s=size)
ax.axis('off')
```

```python
# TODO : automatic point size
size = 5

fig,axes = plt.subplots(2,3,figsize=(20,10))

for i,sample_method in enumerate([ Image.NEAREST, Image.BOX, Image.BILINEAR, Image.HAMMING, Image.BICUBIC, Image.LANCZOS]):
    
    img = Image.open(file_path).convert('RGB')

    converter = ImageEnhance.Color(img)
    new_img = converter.enhance(1.6)

    new_img = new_img.resize(target_size,sample_method) # Image.NEAREST Image.BOX Image.BILINEAR Image.HAMMING Image.BICUBIC Image.LANCZOS
    new_img_ary = np.asarray(new_img) 
    
    
    ax=axes[i//3][i%3]
    ax.set_aspect('equal', adjustable='box')
    ax.scatter([i for i in range(new_img_ary.shape[1])] * new_img_ary.shape[0],
               [[i]*new_img_ary.shape[1] for i in range(new_img_ary.shape[0]-1,0-1,-1)],
               c=list(map(rgb_to_hex,new_img_ary.reshape(-1,3))),s=size)
    ax.axis('off')
    
```
